```python
url = f'https://www.olympedia.org/athletes/1'

reponse = requests.get(url)
soup = BeautifulSoup(reponse.content, 'html.parser')
    
    
table = soup.find('table')
rows = table.find_all('tr')


data = {}

for row in rows:
    
    keys = row.find('th')
    keys = keys.getText()
    
    
    vals = row.find('td')
    vals = vals.getText()
    
    
    data[keys] = f'{vals}'

```

```python
athletes = []

for i in range(1,120000,10):

    try:
        url = f'https://www.olympedia.org/athletes/{i}'

        reponse = requests.get(url)
        soup = BeautifulSoup(reponse.content, 'html.parser')
            
            
        table = soup.find('table')
        rows = table.find_all('tr')


        data = {}

        for row in rows:
            
            keys = row.find('th')
            keys = keys.getText()
            
            
            vals = row.find('td')
            vals = vals.getText()
            
            
            data[keys] = f'{vals}'
            
        if i % 1000 == 0:
            print(i)
            
    except Exception as e:
        print(e)
        
    athletes.append(data)
```
